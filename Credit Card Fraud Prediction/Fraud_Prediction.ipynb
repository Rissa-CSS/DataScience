# **Machine Learning Models**

**Preparing data set for models**

```python
# Seperating features from target
X= df.drop('Class',axis=1)
y= df['Class']
```

```python
#Using SMOTE (Synthetic Minority Over-sampling Technique) to help with target imbalance
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X, y = smote.fit_resample(X, y)
```

```python
#20/80 Test/Train split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

#Initialize and fit model
model_lg = LogisticRegression()
model_lg.fit(x_train,y_train)

# Evaluate its performance
lg_scores = cross_val_score(model_lg, x_train, y_train, cv=10)
lg_mean_score = round(lg_scores.mean(), 4)
lg_mean_score
```

**Linear Discriminant Analysis (LDA)**

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

#Initialize and fit model
model_ld = LinearDiscriminantAnalysis()
model_ld.fit(x_train, y_train)

# Evaluate its performance
ld_scores = cross_val_score(model_ld, x_train, y_train, cv=10)
ld_mean_score = round(ld_scores.mean(), 4)
ld_mean_score
```

**Gaussian Naive Bayes (GNB)**

```python
from sklearn.naive_bayes import GaussianNB

#Initialize and fit model
model_gnb = GaussianNB()
model_gnb.fit(x_train, y_train)

# Evaluate its performance
gnd_score = cross_val_score(model_gnb, x_train, y_train, cv=10)
gnb_mean_score = round(gnd_score.mean(),4)
gnb_mean_score
```

**Prediction with the Logistic Regression Model**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Model prediction
y_pred = model_lg.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")

# Calculate classification report
class_report = classification_report(y_test, y_pred)
print(f"Classification Report:\n{class_report}")
```
